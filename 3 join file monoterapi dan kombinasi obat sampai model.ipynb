```python
modeling = pd.read_csv ('data_modeling_bersih.csv')
```

```python
X = pd.read_csv ('data_modeling_bersih.csv')
```

```python
#train features and target
X.drop(labels = ['sinergy_cat','CELL_LINE','COMPOUND_A','COMPOUND_B'], axis = 1, inplace = True)
y = modeling ['sinergy_cat']
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
   X, y,
    test_size=0.3,
    random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
n=30
rf = RandomForestClassifier(n_estimators=n, max_depth=20, n_jobs=16, warm_start=True)
rf.fit(X_train,y_train)
```

```python
from sklearn import metrics
prediction_test = rf.predict(X_test)
# Print the test accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
prediction_train = rf.predict(X_train)
# Print the train accuracy
print (metrics.accuracy_score(y_train, prediction_train))
```

```python
from sklearn.externals import joblib
joblib.dump(rf, 'model_baruok.joblib')
```
